# استفاده از ماوس

باید یک تابع در بدنه برنامه تعریف کرد وحرکات ماوس را در آن برحسب نیاز کنترل کردو خروجی مد نظر را گرفت 

event =نوع پیشامدی است (2بار فشردن کلید چپ-پایین نگه داشتن کلید چپ-رها کردن کلید چپ و.....)

x , y =مختصات ماوس در لحظه وقوع اتفاق

flags = ctrl,alt,shift برخی پرچم های مرتبط مانند فشردن 

param = opencv پارامترهای اضافی مربوط به 

# cv2.setMouseCallback(windowName,function)

windowName = نام پنجره ای که حرکات ماوس در آن اتفاق میوفتد

function =  تابعی که هنگام رخ دادن پیشامد فراخوانی میشود

```python
import numpy as np
import cv2

drawing = False
ersing = False
color = (0,0,0)

def brush (event,x,y,flags,param):
    global drawing,img,color,ersing
    
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        
    elif event == cv2.EVENT_RBUTTONDOWN:
        ersing = True
        
    elif event == cv2.EVENT_RBUTTONUP:
        ersing = False   
        
    elif event == cv2.EVENT_MOUSEMOVE and drawing == True:
        if flags == (cv2.EVENT_FLAG_CTRLKEY + cv2.EVENT_FLAG_LBUTTON ):
            color = (255,0,0)
        elif flags == (cv2.EVENT_FLAG_ALTKEY + cv2.EVENT_FLAG_LBUTTON):
            color = (0,255,0)
        elif flags == (cv2.EVENT_FLAG_SHIFTKEY + cv2.EVENT_FLAG_LBUTTON):
            color = (0,0,255)
        else:
            color = (0,0,0)
        cv2.circle(img,(x,y),5,color,-1)    
    elif event == cv2.EVENT_MOUSEMOVE and ersing == True:
        color = (255,255,255)
        cv2.circle(img,(x,y),5,color,-1)
    
        
img = np.ones((480,640,3),np.uint8)*255
cv2.namedWindow("Paper")
cv2.setMouseCallback("Paper",brush)



while(True):
    key = cv2.waitKey(1)
    if key == ord("s"):
        cv2.imwrite("Painted.jpg",img)
    elif key == 27:
        break
    cv2.imshow("Paper",img)
cv2.destroyAllWindows()

```

# بارگیری تصاویر از اینترنت

urllib کتابخانه
 
urlopenدانلود با تابع 

به آرایه 2 بعدی تبدیل میشودasarray -bytearray باتابع 

imdecode خواندن تصویر از حافظه موقت سیستم

```python
import numpy as np
import cv2
import urllib.request

url = ["http://www.pyimagesearch.com/wp-content/uploads/2015/01/opencv_logo.png"]
#print(url)
def url_to_image(url):
    don = urllib.request.urlopen(url)
    image = np.asarray(bytearray(don.read()),np.uint8)
    image = cv2.imdecode(image,cv2.IMREAD_COLOR)
    return image
for url in url:
    print ("Downloading %s"%(url))
    image = url_to_image(url)
    cv2.imshow("Image",image)
    key = cv2.waitKey(0)
cv2.destroyAllWindows()
```
