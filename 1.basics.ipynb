# ذخیره فیلم به کمک دوربین

save = cv2.VideoWriter("filename",fourCC,fps,frameSize)

```python
import cv2

camera = cv2.VideoCapture(0)
fourCC =cv2.VideoWriter_fourcc(*"XVID")
save = cv2.VideoWriter("output.mp4",fourCC,25.0,(640,480))
recording = False

if camera.isOpened() == False:
    camera.open(0)
    print("please open your camera")
    
else:
    print("camera is detected")
while(camera.isOpened()):
    k = cv2.waitKey(1)
    if k == 27:
        break
    elif k == ord('r'):
        recording = not(recording)
    _,frame = camera.read()
    cv2.imshow("video",frame)
    if recording == True:
        save.write(frame)
    
camera.release()
save.release()
cv2.destroyAllWindows()

```

# دسترسی به ویژگی های دوربین

```python
import cv2
import time

cam = cv2.VideoCapture(0)
num_frames = 120

start = time.time()
for i in range(0,num_frames):
    ret,frame =cam.read() 
end = time.time()

seconds = end - start
fps = num_frames / seconds

print("FPS is :(%d )" % int(fps))
print("resolution is :(%d ,%d)" %(cam.get(3),cam.get(4)))
cam.release()
```
